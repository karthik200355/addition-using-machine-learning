```python
#store feature matrix in x and response in y 
x=data[['x','y']]
y=data['sum']
```

```python
#train and test split
#1.split data into 2 parts a training data set and a testing dataset
#2.train model on training set
#3.test model on testing set
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
   x, y, test_size=0.33,random_state=42)
```

```python
 x_train
```

```python
#import and train the model 
from sklearn.linear_model import LinearRegression 
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#chech model prediction performane
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
#compare the result
y_pred=model.predict(x_test)
y_pred
```

```python
df=pd.DataFrame({'actual':y_test ,'predict':y_pred})
df
```

```python
from tkinter import *
import joblib

def show_entry_fields():
    try:
        p1 = float(e1.get())
        p2 = float(e2.get())
        
        model = joblib.load('model_joblib')
        result = model.predict([[p1, p2]])
        
        label_sum = Label(master, text='Predicted Sum is:')
        label_sum.grid(row=4, column=0)
        
        label_result = Label(master, text=str(result[0]))
        label_result.grid(row=4, column=1)
        
    except Exception as e:
        error_label = Label(master, text=f"Error: {e}", fg="red")
        error_label.grid(row=5, columnspan=2)

master = Tk()
master.title("Karthik")

label_title = Label(master, text="Addition of Two Numbers Using ML", bg='black', fg='white')
label_title.grid(row=0, columnspan=2)

label1 = Label(master, text="Enter first number")
label1.grid(row=1, column=0)

label2 = Label(master, text="Enter second number")
label2.grid(row=2, column=0)

e1 = Entry(master)
e2 = Entry(master)

e1.grid(row=1, column=1)
e2.grid(row=2, column=1)

Button(master, text='Predict', command=show_entry_fields).grid(row=3, columnspan=2)

mainloop()
```
